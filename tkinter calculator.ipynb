```python
from tkinter import *
class Demo(Frame):
    def __init__(self,m):
        super().__init__(m)
        
        self.v1=0
        self.v2=0
        self.opr=''
        
        self.t1=Entry(self)
        self.b1=Button(self,text='1',width='4',height='1',font=('algerian',14),command=self.one)
        self.b2=Button(self,text='2',width='4',height='1',font=('algerian',14),command=self.two)
        self.b3=Button(self,text='3',width='4',height='1',font=('algerian',14),command=self.three)
        self.b4=Button(self,text='4',width='4',height='1',font=('algerian',14),command=self.four)
        self.b5=Button(self,text='5',width='4',height='1',font=('algerian',14),command=self.five)
        self.b6=Button(self,text='6',width='4',height='1',font=('algerian',14),command=self.six)
        self.b7=Button(self,text='7',width='4',height='1',font=('algerian',14),command=self.seven)
        self.b8=Button(self,text='8',width='4',height='1',font=('algerian',14),command=self.eight)
        self.b9=Button(self,text='9',width='4',height='1',font=('algerian',14),command=self.nine)
        self.b10=Button(self,text='0',width='6',height='1',font=('algerian',14),command=self.zero)
        
        self.b11=Button(self,text='+',width='4',height='1',font=('algerian',14),command=self.plus)
        self.b12=Button(self,text='-',width='4',height='1',font=('algerian',14),command=self.minus)
        self.b13=Button(self,text='*',width='4',height='1',font=('algerian',14),command=self.multiply)
        self.b14=Button(self,text='/',width='4',height='1',font=('algerian',14),command=self.divide)
        self.b15=Button(self,text='=',width='4',height='1',font=('algerian',14),command=self.euqalto)
        self.b16=Button(self,text='.',width='4',height='1',font=('algerian',14),command=self.dot)
        self.b17=Button(self,text='C',width='4',height='1',font=('algerian',14),command=self.clear)
        self.b18=Button(self,text='B',width='4',height='1',font=('algerian',14),command=self.back)
        
        
        self.rowconfigure(index=0,pad=3)
        self.rowconfigure(index=1,pad=3)
        self.rowconfigure(index=2,pad=3)
        self.rowconfigure(index=3,pad=3)
        self.rowconfigure(index=4,pad=3)
        
        self.columnconfigure(index=0,pad=3)
        self.columnconfigure(index=1,pad=3)
        self.columnconfigure(index=2,pad=3)
        self.columnconfigure(index=3,pad=3)
        
        self.t1.grid(columnspan=4)
        
        self.b7.grid(row=1,column=0)
        self.b8.grid(row=1,column=1)
        self.b9.grid(row=1,column=2)
        
        self.b4.grid(row=2,column=0)
        self.b5.grid(row=2,column=1)
        self.b6.grid(row=2,column=2)
        
        self.b1.grid(row=3,column=0)
        self.b2.grid(row=3,column=1)
        self.b3.grid(row=3,column=2)
        
        self.b10.grid(row=4)
        
        self.b11.grid(row=1,column=4)
        self.b12.grid(row=2,column=4)
        self.b13.grid(row=3,column=4)
        self.b14.grid(row=4,column=4)
        self.b15.grid(row=5,column=4)
        
        self.b16.grid(row=5,column=0)
        self.b17.grid(row=5,column=1)
        self.b18.grid(row=5,column=2)
        
        self.pack()
        
    def one(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'1'
        self.t1.insert(0,v)
    def two(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'2'
        self.t1.insert(0,v)
    def three(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'3'
        self.t1.insert(0,v)
    def four(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'4'
        self.t1.insert(0,v)
    def five(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'5'
        self.t1.insert(0,v)
    def six(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'6'
        self.t1.insert(0,v)
    def seven(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'7'
        self.t1.insert(0,v)
    def eight(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'8'
        self.t1.insert(0,v)
    def nine(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'9'
        self.t1.insert(0,v)
    def zero(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'0'
        self.t1.insert(0,v)
    def dot(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        v=v+'.'
        self.t1.insert(0,v)
    def plus(self):
        self.v1=int(self.t1.get())
        self.opr='+'
        self.t1.delete(0,'end') 
    def minus(self):
        self.v1=int(self.t1.get())
        self.opr='-'
        self.t1.delete(0,'end')
    def multiply(self):
        self.v1=int(self.t1.get())
        self.opr='*'
        self.t1.delete(0,'end')
    def divide(self):
        self.v1=int(self.t1.get())
        self.opr='/'
        self.t1.delete(0,'end')
    
            
    def euqalto(self):
        self.v2=int(self.t1.get())
        if self.opr=='+':
            self.v1+=self.v2
        elif self.opr=='-':
            self.v1-=self.v2
        elif self.opr=='*':
            self.v1*=self.v2
        elif self.opr=='/':
            self.v1/=self.v2
            
        self.t1.delete(0,'end')
        self.t1.insert(0,self.v1)
    def clear(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
    def back(self):
        v=self.t1.get()
        self.t1.delete(0,'end')
        s=len(v)-1
        v=v[:s]
        self.t1.insert(0,v)
    
             
        
root=Tk()
obj=Demo(root)
root.title('Calculator')
root.geometry('300x300')
root.mainloop()
```
